```python
import os
import sys
import pandas as pd
import openpyxl
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
xfiles = [f for f in os.listdir() if f.endswith("xlsx")]
xfiles
factor_file_name = [f for f in xfiles if f.startswith("Cent")][0]
factor_file_name
```

```python
#with open(factor_file_name, mode='r') as file:
df = pd.read_excel(factor_file_name, sheet_name='Century of Factor Premia', skiprows=18 )
print(type(df))
df.columns
```

```python
df['dt'] = pd.to_datetime(df["Date"])
df.head()
df.columns
ax = df.iloc[1000:].plot(x="dt", y='US Stock Selection Value', figsize=[25,10], marker=".")
plt.rc('xtick',labelsize=38)
plt.rc('ytick',labelsize=42)
plt.axhline(y=0)
```
